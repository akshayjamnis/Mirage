```python
import random
import pandas as pd

# 1️⃣ Start with a fixed seed for reproducibility
random.seed(42)

# 2️⃣ Define possible values
GENDERS = ["Male", "Female"]

# 3️⃣ Function to generate one random instance
def generate_instance():
    # Basic demographics
    age = random.randint(15, 80)
    gender = random.choice(GENDERS)

    # Mental health features (0–10 scale)
    stress = random.randint(0, 10)
    anxiety = random.randint(0, 10)
    depression = random.randint(0, 10)
    sleep_quality = random.randint(0, 10)   # higher = better
    social_support = random.randint(0, 10)  # higher = better
    physical_activity = random.randint(0, 10)
    work_pressure = random.randint(0, 10)

    # 4️⃣ Algorithmic logic for mental state
    risk_score = (
        stress * 1.5 +
        anxiety * 1.7 +
        depression * 2.0 -
        sleep_quality * 1.2 -
        social_support * 1.0 -
        physical_activity * 0.8 +
        work_pressure * 1.3
    )

    # Normalize logic into categories
    if risk_score < 10:
        mental_state = "Healthy"
    elif 10 <= risk_score < 25:
        mental_state = "Mild"
    elif 25 <= risk_score < 40:
        mental_state = "Moderate"
    else:
        mental_state = "Severe"

    # 5️⃣ Return feature status as a dictionary
    return {
        "age": age,
        "gender": gender,
        "stress": stress,
        "anxiety": anxiety,
        "depression": depression,
        "sleep_quality": sleep_quality,
        "social_support": social_support,
        "physical_activity": physical_activity,
        "work_pressure": work_pressure,
        "mental_state": mental_state
    }

# 6️⃣ Generate 1000 instances
data = [generate_instance() for _ in range(1000)]

# 7️⃣ Create DataFrame
df = pd.DataFrame(data)

# 8️⃣ Save to CSV
csv_path = "mental_health_dataset.csv"
df.to_csv(csv_path, index=False)

# 9️⃣ Show head
print(df.head())

print(f"\nDataset saved to: {csv_path}")
```
